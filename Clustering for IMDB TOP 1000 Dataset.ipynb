```python
df['Gross'] = df['Gross'].astype(str).str.replace('$', '')  # Remove dollar signs
df['Gross'] = df['Gross'].str.replace(',', '')  # Remove commas
df['Gross'] = df['Gross'].str.strip()  # Remove leading/trailing spaces
```

```python
def convert_suffix(value):
    try:
        value = value.strip().upper()
        if value.endswith('M'):
            return float(value[:-1]) * 1_000_000
        elif value.endswith('B'):
            return float(value[:-1]) * 1_000_000_000
        else:
            return float(value)
    except:
        return np.nan
```

```python
df['Gross'] = df['Gross'].apply(convert_suffix)
```

```python
print("Number of NaN values:", df['Gross'].isna().sum())
print("\nFirst few converted values:")
print(df['Gross'].head())
```

```python
# Replace nulls with median
df['Gross'] = df['Gross'].fillna(df['Gross'].median())
```

```python
df['Gross'] = df['Gross'].astype('int64')  # Converts to int
```

```python
df.loc[:, 'Meta_score'] = df['Meta_score'].fillna(df['Meta_score'].mean()) #replacing null value with mean
```

```python
df['Meta_score'] = df['Meta_score'].astype('int64')  # Converts to int
```

```python
# checking null after imputation
df.isnull().sum()
```

```python
# Drops multiple columns
df = df.drop(['Poster_Link', 'Overview'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

# Specify the indices of the columns you want to scale
column_indices = [6, 13]  # Replace with the indices of the columns you want to scale

# Select the columns by their indices
columns_to_scale = df.columns[column_indices]

# Apply standard scaling to these selected columns only
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[columns_to_scale])

# Replace the specified columns in the original DataFrame with the scaled values
df[columns_to_scale] = scaled_data
```

```python
# Calculate elbow curve
from sklearn.cluster import KMeans

inertias = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(scaled_data)
    inertias.append(kmeans.inertia_)
```

```python
# Plot elbow curve
plt.figure(figsize=(12, 5))
plt.plot(K, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method')
```

```python
# Step 1: Fit K-Means Clustering
from sklearn.cluster import KMeans
optimal_k = 4  # Choose the number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(scaled_data)
```

```python
# Get cluster labels
clusters = kmeans.labels_
```

```python
plt.figure(figsize=(10, 7))
# Scatter plot for the data points
scatter = plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', edgecolor='k', label='Data Points')

# Scatter plot for the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red', marker='X', label='Centroids')

plt.title('K-Means Clustering (k=5) Visualization')
plt.xlabel('Meta_score')
plt.ylabel('Gross')

# Explicitly call the legend
plt.legend()

plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Assuming you have your features in X (the data points) and clusters are already assigned

# Compute the silhouette score for the clustering
sil_score = silhouette_score(scaled_data, clusters)
print(f"Silhouette Score for the K-means clustering: {sil_score}")

# Count number of customers in each cluster
cluster_counts = pd.Series(clusters).value_counts().sort_index()
print("\nNumber of customers in each cluster:")
for cluster, count in cluster_counts.items():
    print(f"Cluster {cluster}: {count} customers")
```

```python
import scipy.cluster.hierarchy as sch 

plt.figure(figsize=(10, 7))
dendrogram=sch.dendrogram(sch.linkage(scaled_data,method  =  'ward'))
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Meta_score')
plt.ylabel('Gross')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, metric = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(scaled_data)
```

```python
# Calculate silhouette score
sil_score = silhouette_score(scaled_data, y_hc)
print(f"Silhouette Score for the AgglomerativeClustering: {sil_score}")

# Plot
plt.figure(figsize=(10, 7))
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=y_hc, cmap='viridis', edgecolor='k')
plt.title(f'Hierarchical Clustering (k=5)')
plt.xlabel('Meta_score')
plt.ylabel('Gross')
plt.legend(['Data Points'])
plt.show()
```
